```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score, normalized_mutual_info_score
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import SpectralEmbedding
from sklearn.metrics.pairwise import cosine_similarity
import networkx as nx

# Initialize an empty graph
G = nx.Graph()
# Step 1: Build a graph of categorical values using the Cosine similarity measure
similarity_matrix = cosine_similarity(zoo_df)
    
    # Add nodes to the graph
n_samples, n_features = zoo_df.shape
for i in range(n_samples):
    for j in range(n_features):
        G.add_node((i, j))
    
    # Add edges to the graph based on cosine similarity
for i in range(n_samples):
    for j in range(i+1, n_samples):
        for k in range(n_features):
            for l in range(k+1, n_features):
                sim = similarity_matrix[i, j]
                G.add_edge((i, k), (j, l), weight=sim)

# Step 2: Get a representation matrix of categorical values f(v) by Spectral Embedding
embedding_model = SpectralEmbedding(n_components=10)
spectral_embedding = embedding_model.fit_transform(similarity_matrix)
# Visualize the embedding space
plt.scatter(spectral_embedding[:, 0], spectral_embedding[:, 1])
plt.title('Embedding Space')
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.show()
    

# Step 3: Get a representation matrix of objects g(X) by concatenating numeric features
scaler = StandardScaler()
numeric_features = zoo_df[['legs']]
scaled_numeric_features = scaler.fit_transform(numeric_features)
object_representation = np.concatenate([spectral_embedding, scaled_numeric_features], axis=1)

# Step 4: Compute clustering using K-means
kmeans = KMeans(n_clusters=7, random_state=42)
cluster_labels = kmeans.fit_predict(object_representation)
print(cluster_labels)
# # Evaluation using performance metrics
# true_labels = y
# ari_score = adjusted_rand_score(true_labels, cluster_labels)
# nmi_score = normalized_mutual_info_score(true_labels, cluster_labels)

# print("Adjusted Rand Index (ARI):", ari_score)
# print("Normalized Mutual Information (NMI):", nmi_score)
```

```python
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics.cluster import adjusted_rand_score, normalized_mutual_info_score
from sklearn.metrics import fowlkes_mallows_score
from kmodes.kmodes import KModes
import pandas as pd
from sklearn.impute import SimpleImputer

# data (as pandas dataframes) 
zoo_df = zoo.data.features 
y = zoo.data.targets.values
y = y.ravel()  # Reshape y to make it 1D

# Perform clustering using KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(zoo_df)
kmeans_labels = kmeans.labels_

# Perform clustering using Hiearchical Clustering
agg_clustering = AgglomerativeClustering(n_clusters=3)
agg_clustering.fit(zoo_df)
agg_labels = agg_clustering.labels_

# Perform clustering using KModes
km = KModes(n_clusters=3, init='Huang', n_init=5, verbose=0)
km.fit_predict(zoo_df)
km_labels = km.labels_

# Evaluate clustering performance using adjusted Rand index (ARI), normalized mutual information (NMI), and Fowlkes-Mallows index (FMI)
ari_kmeans_score = adjusted_rand_score(y, kmeans_labels)
nmi_kmeans_score = normalized_mutual_info_score(y, kmeans_labels)
fmi_kmeans_score = fowlkes_mallows_score(y, kmeans_labels)
ari_agg_score = adjusted_rand_score(y, agg_labels)
nmi_agg_score = normalized_mutual_info_score(y, agg_labels)
fmi_agg_score = fowlkes_mallows_score(y, agg_labels)
ari_km_score = adjusted_rand_score(y, km_labels)
nmi_km_score = normalized_mutual_info_score(y, km_labels)
fmi_km_score = fowlkes_mallows_score(y, km_labels)

# Print the scores
print("Adjusted Rand Index (ARI) with K-Means:", ari_kmeans_score)
print("Normalized Mutual Information (NMI) with K-Means:", nmi_kmeans_score)
print("Fowlkes-Mallows Index (FMI) with K-Means:", fmi_kmeans_score)
print("")
print("Adjusted Rand Index (ARI) with Hierachical Clustering:", ari_agg_score)
print("Normalized Mutual Information (NMI) Hierachical Clustering:", nmi_agg_score)
print("Fowlkes-Mallows Index (FMI) Hierachical Clustering:", fmi_agg_score)
print("")
print("Adjusted Rand Index (ARI) with KModes:", ari_km_score)
print("Normalized Mutual Information (NMI) with KModes:", nmi_km_score)
print("Fowlkes-Mallows Index (FMI) with KModes:", fmi_km_score)
```
